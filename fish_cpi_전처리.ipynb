```python
import pandas as pd
```

```python
data = pd.read_csv('/content/drive/MyDrive/BigContest/DATA/cpi/fish_cpi.csv', encoding='cp949')
```

```python
data.head(3)
```

```python
data.drop('Unnamed: 0',axis=1, inplace = True)
data.columns = [month.replace('.','-').replace(' ','').replace('월','') for month in data.columns]
```

```python
# data.info()
```

```python
data = data.transpose()
data_0 = data.iloc[0,:]
data_1 = data.iloc[1:,:]
data_1 = data_1.T
```

```python
data_1.columns = pd.to_datetime(data_1.columns)
data_1.head()
```

```python
final_df = pd.concat([data_0, data_1], axis=1)
```

```python
final_df.to_csv('/content/drive/MyDrive/final.csv')
```
