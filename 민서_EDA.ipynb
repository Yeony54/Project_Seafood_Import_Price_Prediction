## P_PURPOSE
수입용도

```python
df_train.P_PURPOSE.describe()
```

5가지의 수입용도가 있다

```python
print(df_train['P_PURPOSE'].value_counts())
```

얘네는 1 - 5 로 카테고리화 할 수 있을 듯 (코드는 따로 보관)

```python
def get_purpose(row):
    if row.P_PURPOSE == "판매용":
        return 1
    elif row.P_PURPOSE == "자사제품제조용":
        return 2
    elif row.P_PURPOSE == "외화획득용 원료":
        return 3
    elif row.P_PURPOSE == "합작":
        return 4
    elif row.P_PURPOSE == "반송품(기타)":
        return 5
```

```python
df_train['P_PURPOSE'] = df_train.apply(lambda x: get_purpose(x), axis=1)
```
